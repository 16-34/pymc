```python
from mcpi.minecraft import Minecraft as MC
import cv2
import matplotlib.pyplot as plt
from woolcolor import WoolColor
```

```python
mc = MC.create()
mc.postToChat("阿巴阿巴")
mc.postToChat("天才喜欢睡觉")
```

```python
x0, y0, z0 = mc.player.getTilePos()
x0, y0, z0
```

```python
img_bgr = cv2.imread('1.jpeg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
img_rgb.shape
```

```python
wc = WoolColor()
```

```python
h, w, d = img_rgb.shape
clrs = wc.lab_track(img_rgb)

sta_x = x0 - w//2
sta_z = z0 - h//2

for y in range(h):
    for x in range(w):
        mc.setBlock(sta_x+x, y0-1, sta_z+y, 35, clrs.item(y, x))
```

```python
mc.setBlocks(sta_x, y0-1, sta_z, sta_x+w-1, y0-1, sta_z+h-1, 2)
```
